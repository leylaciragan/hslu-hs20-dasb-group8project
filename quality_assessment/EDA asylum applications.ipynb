### Exploratory Data Analysis 
#### Asylum Applications in Switzerland
> This section generates a profile no our main data set. This profile was used to get a general overview on our data, e.g. for the first steps of our Exploratory Data Analysis like findings missing values, interpreting zeros, getting maximum and minimum values etc.
> If you toggle details for each variable, you can find more detailed information on it quantiles, mean, skewness etc.

```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
applications = pd.read_csv("data/AsylgesuchePerNation1986.csv", sep=";")
applications.head()
```

```python
# save a profile report to html
profile = applications.profile_report(minimal=True)

# Print profile here
profile
```

```python
# Save report to html
profile.to_file(output_file="applications-profile.html")
```
